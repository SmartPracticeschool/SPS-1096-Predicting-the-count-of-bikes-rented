## Data Preprocessing

2.Removing Outliers

```python
#We have removed Null values previously (above)

#Now,
#Dropping irrelevent columns

df = df.drop(['instant','dteday'], axis=1)
df.head()
```

### Then split the data into dependent and independent variables
###### can use iloc to read the columns....like

```python
x = df.iloc[:,:15]   #input/independent values
x.head()
```

```python
y = df.iloc[:,-1:]   #output/dependent values
y.head()
```

## Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

## Applying Linear Regression / FeatureScaling

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
linr = LinearRegression()

linr.fit(x_train, y_train)
```

```python
x_train
```

## Predictions

```python
pred = linr.predict(x_test)
pred
```

```python
plt.scatter(y_test, pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```
